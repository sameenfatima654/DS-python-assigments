```python
import pandas as pd
```

```python
data= pd.read_csv("Pokemon.csv")
data.head()
```

```python
attack_mean = data["Attack"].mean()
attack_mean
```

```python
#we have to add a col after legendary which will tell us if the attack is low or high wrt to mean value which is 79.00125
```

```python
def set_attack(val):
    if val < attack_mean:
        return "Attack Low"
    else:
        return "Attack High"
```

```python
data["Attack_high_low"] = data["Attack"].apply(set_attack)
data
```

### ASSIGNMENT

###### Add 2 cols named "HP_high_low" and "speed_high_low" corresponding to cols "speed" and "HP" 

```python
hp_mean = data["HP"].mean()
hp_mean
```

```python
def set_hp(val):
    if val < hp_mean:
        return "HP Low"
    else:
        return "HP High"
```

```python
data["HP_high_low"] = data["HP"].apply(set_hp)
data
```

```python
speed_mean = data["Speed"].mean()
speed_mean
```

```python
def set_speed(val):
    if val < speed_mean:
        return "Speed Low"
    else:
        return "Speed High"
```

```python
data["Speed_high_low"] = data["Speed"].apply(set_speed)
data
```
